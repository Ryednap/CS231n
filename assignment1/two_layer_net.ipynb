## Inline Question 1:

We've only asked you to implement ReLU, but there are a number of different activation functions that one could use in neural networks, each with its pros and cons. In particular, an issue commonly seen with activation functions is getting zero (or close to zero) gradient flow during backpropagation. Which of the following activation functions have this problem? If you consider these functions in the one-dimensional case, what types of input would lead to this behavior?
1. Sigmoid
2. ReLU
3. Leaky ReLU

## Answer:
* Sigmoid: Has saturation endpoints for larger and smaller inputs which leads to vanishing gradient. Only active for inputs close to origin.
* ReLU: Kills the gradient for negative inputs completely leading to dead neurons. Only positive input receives gradient update. Alo not zero centered meaning that in deep-layers input will be close to zero as the result the gradient update to front-layers will be close to zero.
* Leaky ReLU: No zero centred hence the same problem with deep-networks.

**Note**: With good learning design any model can have decent training accuracy it is the test-time accuracy that we really care about. So either we decrease the train accuracy (which has over-fit) or increase the test-accuracy.
